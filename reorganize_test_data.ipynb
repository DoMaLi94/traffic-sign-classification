```python
import os
import shutil
import pandas as pd

DATA_DIR = "./data/gtsrb-german-traffic-sign/"
```

```python
file_path = os.path.join(DATA_DIR, 'Test.csv')
data = pd.read_csv(file_path)
df = data.loc[:, ['Path', 'ClassId']]
dest_paths = []

for path, classid in df.to_numpy():
    src_path = os.path.join(DATA_DIR, path)
    
    path = os.path.join('Test', str(classid), os.path.basename(path))
    dest_path = os.path.join(DATA_DIR, path)

    print("Source:", src_path, "| Destination:", dest_path)

    # os.makedirs(dest_path, exist_ok=True)
    # shutil.move(src_path, dest_path)

    dest_paths.append(path)

data['Path'] = dest_paths
    
```
